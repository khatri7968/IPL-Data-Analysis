**Q:4** - No of matches played in each season?

```python
season_df = ipl_df.groupby('Season')[["ID"]].count()
season_df = season_df.sort_values('Season', ascending = False).reset_index()
season_df.rename(columns = {'ID':'Matches', 'Season':'Year'}, inplace = True)
```

```python
season_df
```

```python
plt.figure(figsize = (20, 10))
plt.grid()
plt.title("Matches played in each season")
sns.barplot(x='Matches', y='Year', data = season_df);
```

By looking above figure we can say that highest number of matches played in 2013 and then in 2012 and then in 2011. 

**Q:5** - Nunber of matches loss by each team?

```python
loss_matches = []
for i in ipl_df.Team1:
    for j in ipl_df.Team2:
        if i not in ipl_df.WinningTeam:
            loss_matches.append(j)
        else:
            loss_matches.append(i)
```

```python
from collections import Counter
number_of_matches = Counter(loss_matches)
```

```python
loss_matches_df = pd.DataFrame(list(number_of_matches.items()),columns = ['teams','matches']) 
loss_matches_df = loss_matches_df.sort_values('matches', ascending = False)
```

```python
loss_matches_df
```

```python
plt.figure(figsize = (20, 10))
plt.grid()
plt.title("Matches loss by teams in each season")
sns.barplot(x='matches', y='teams', data = loss_matches_df);
```

#### Q1: what is most perferrable b/w bat and bowl after winning toss

#### Q2: Which decision has proved most benificial?

```python
field_df = ipl_df.loc[(ipl_df['TossWinner'] == ipl_df['WinningTeam']) & (ipl_df['TossDecision'] == 'field'), ['ID', 'WinningTeam','TossDecision']]
```

```python
field_df.WinningTeam.count()
```

```python
bat_df = ipl_df.loc[(ipl_df['TossWinner'] == ipl_df['WinningTeam']) & (ipl_df['TossDecision'] == 'bat'), ['ID', 'WinningTeam','TossDecision']]
```

```python
bat_df.WinningTeam.count()
```

```python
frames = [bat_df, field_df]
result_df = pd.concat(frames)
result_df = result_df.groupby('TossDecision')[['ID']].count()
result_df
```

```python

result_df = result_df.sort_values('ID').reset_index()
result_df.rename(columns={'ID':'Total','TossDecision':'TossDecision'},inplace=True)
result_df
```

```python
plt.figure(figsize=(10,10))
plt.title("Decision Success",fontweight='bold',fontsize=30)
plt.xlabel('Decision',fontweight='bold',fontsize=30)
plt.ylabel('Total',fontweight='bold',fontsize=30)
plt.tick_params(labelsize=20)
plt.bar(result_df.TossDecision, result_df.Total, color=['#00FF00','#00FF00']);
plt.legend(['Decision Proved Right']);
```
